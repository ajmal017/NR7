```python
from ib_insync import *
util.startLoop()

ib = IB()
ib.connect('127.0.0.1', 7496, clientId=14)
```

```python
contract = Stock('TCS', 'NSE', 'INR')

ib.reqHeadTimeStamp(contract, whatToShow='TRADES', useRTH=True)
```

```python
bars = ib.reqHistoricalData(
        contract,
        endDateTime='',
        durationStr='6 D',
        barSizeSetting='5 mins',
        whatToShow='TRADES',
        useRTH=True,
        formatDate=1)
```

```python
df = util.df(bars)
df = df.dropna()
df['date'] = pd.to_datetime(df['date'])
df= df.set_index('date')
min_df = (df.resample('10min', base=5)
               .agg({'open': 'first', 'high': 'max', 'low': 'min', 'close': 'last', 'volume':'mean'}))

daily_df = (df.resample('1 D')
               .agg({'open': 'first', 'high': 'max', 'low': 'min', 'close': 'last', 'volume':'mean'}))
```

```python
daily_df= daily_df.dropna()
daily_df.index
```

```python
daily_df['Pivot'] = (daily_df['high'].shift(1) + daily_df['low'].shift(1) + daily_df['close'].shift(1))/3
daily_df['R1'] = 2*daily_df['Pivot'].shift(1) - daily_df['low'].shift(1)
daily_df =daily_df.dropna()                                            
display(daily_df)
```

```python
for t in df.index:
    if (t.date() in daily_df.index):
        print(daily_df.loc[t.date().strftime('%Y-%m-%d')])
```

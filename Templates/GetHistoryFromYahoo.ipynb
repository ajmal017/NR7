```python
import pandas_datareader as web
import ta, datetime
```

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2020, 1, 1)
df = web.DataReader("^NSEI", 'yahoo', start, end)
#df1 = web.DataReader("TCS.NS", 'yahoo', start, end)
df.head()
```

```python
ma40 = ta.trend.SMAIndicator(df['Close'], 40).sma_indicator()
df['sma40'] = ma40.sort_values(ascending=True)
df.tail()
```

```python
df['Long'] = df['Close'] > df['sma40']
df.tail()
```

```python
df['DReturns'] = df['Close'] - df['Open']
df.tail()
```

```python
df['Long'] = df.Long.shift(1)
df.tail()
```

```python
df_returns = df[df['Long']==True]
df_returns.head()
```

```python
df_returns.DReturns.sum()
```

```python
df['Short'] = df['Close'] < df['sma40']
df['Short'] = df.Short.shift(1)
df['SReturns'] = df['Open'] - df['Close']
df_sreturns = df[df['Short']==True]
df_sreturns.head()
```

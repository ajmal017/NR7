```python
import pandas_datareader as web
import ta, datetime
import numpy as np
```

```python
start = datetime.datetime(2018, 1, 1)
end = datetime.datetime(2020, 9, 1)
df = web.DataReader("^NSEBANK", 'yahoo', start, end)
#df = web.DataReader("RELIANCE.NS", 'yahoo', start, end)
```

```python
ma40 = ta.trend.SMAIndicator(df['Close'], 40).sma_indicator()
ma2 = ta.trend.SMAIndicator(df['Close'], 2).sma_indicator()
ma5 = ta.trend.SMAIndicator(df['Close'], 5).sma_indicator()
df['highc50'] = df['Adj Close'][-50:].idxmax() 
df['lowc50'] = df['Adj Close'][-50:].idxmin()
df['sma40'] = ma40.sort_values(ascending=True)
df['ma2'] = ma2.sort_values(ascending=True)
df['ma5'] = ma5.sort_values(ascending=True)

#drop 50 records 
df =  df.iloc[30:]
df.dropna()
```

```python
df['Cond3'] = np.where(df['highc50'] < df['lowc50'],1,-1)
df['Cond1'] = np.where(df['Close'] > df['sma40'],1,-1)
df['Cond2'] = np.where(df['ma2']<df['ma5'],1,-1)
```

```python
df['Long'] = df['Cond2'] + df['Cond3']
```

```python
df['DReturns'] = (df['Close'] - df['Open'])/df['Open']
df['SReturns'] = (df['Open'] - df['Close'])/df['Close']
df['Long'] = df['Long'].shift(1)
```

```python
import matplotlib as plt
%matplotlib inline
dreturns = df.loc[df.Long>0]
dailyr = dreturns['DReturns']
(dailyr+1).cumprod().plot(figsize=(10,7),grid=True)
```

```python
sreturns = df.loc[df.Long<0]
dailysr = sreturns['SReturns']
(dailysr+1).cumprod().plot(figsize=(10,7),grid=True)
```

```python
import pyfolio as pf
pf.create_simple_tear_sheet(dailysr)
```
